```python
import pandas as pd
import numpy as np
import math
import glob
import csv
```

```python
stocks = {}
count = 1
for files in glob.glob('C:\\Users\\personal\\Desktop\\MSIT\\IT\\DataScience_2019501026\\Data Analytics and Data Visualisation\\Assignments\\Vector Graphics\\Sensex_Data\\*.csv'):
    file = open(files, 'r')
    
    if not file.read(1):
        stocks[str(count)] = [0]
        count = count+1
        continue
    data = pd.read_csv(file)
    stocks[str(count)] = list(data["Close Price"])
    count = count + 1
    file.close()
stocks.keys()
```

```python
# To perform the correlation, the data should have equal sized lists
max1 = 0
for each in stocks.values():
    max1 = max(max1, len(each))

for each in stocks.values():
    each += [0.0] * (max1 - len(each))
```

```python
from scipy import interpolate
corr_data = []
for i in stocks.keys():
    for j in stocks.keys():
        corr = np.corrcoef(stocks.get(i), stocks.get(j)) [0][1]
        if np.isnan(corr):
            corr_data.append(0)
        else:
            corr_data.append(corr)
corr_data
```

```python
with open('correlationdata.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Company1","Company2","Value"])
    i = 1
    for x in stocks.keys():
        j = 1
        for y in stocks.keys():
            corr_value = np.corrcoef(stocks.get(x), stocks.get(y)) [0][1]
            if np.isnan(corr_value):
                writer.writerow([i, j, 0])
            else:
                writer.writerow([i, j, corr_value])
            j = j + 1
        i = i+1
print("Correlation Data is in correlationdata.csv file")
```
