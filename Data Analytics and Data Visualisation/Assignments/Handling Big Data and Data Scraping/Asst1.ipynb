```python
import numpy as np
import pandas as pd
```

```python
title_basics_data = pd.read_csv("title_basics_data.tsv", delimiter='\t', low_memory= False)
title_basics_data.head(5)
```

```python
title_ratings_data = pd.read_csv("title_ratings_data.tsv", delimiter = "\t")
title_ratings_data.head(5)
```

```python
rating_data1 = title_ratings_data[title_ratings_data['averageRating'] > 8.0]
rating_data1.head(20)
```

```python
final_ratings_data = rating_data1.sort_values('numVotes', ascending = False)
final_ratings_data.head(20)
```

```python
merged_data = pd.merge(final_ratings_data, title_basics_data, on= 'tconst')
merged_data.head(10)
```

```python
merged_data = merged_data[merged_data['titleType'] == 'movie']
merged_data.head(10)
```

```python
merged_data['originalTitle'].head(20)
```

```python
title_basics_data.dtypes
```

```python
title_basics_data['startYear'] = pd.to_numeric(title_basics_data['startYear'], errors='coerce')
title_basics_data.dtypes
```

```python
year = title_basics_data[(title_basics_data['startYear'] >= 2000) & (title_basics_data['titleType'] == 'movie')]
year.head(10)
```

```python
updated_merged_data = pd.merge(year, title_ratings_data, on= 'tconst')
updated_merged_data.head(10)
```

```python
votes_data = updated_merged_data[updated_merged_data['numVotes'] > 40000]
votes_data.head(10)
```

```python
final_data = votes_data.sort_values('averageRating', ascending = False)
final_data.head(20)
```
